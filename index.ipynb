```python
import pandas as pd
years= [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019,2020]
duration = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
movie_dict = {year: dur for year, dur in zip(years, duration)}
print(movie_dict)
```

```python
import pandas as pd
duration_df = pd.DataFrame({'duration':movie_dict.values()},index=movie_dict.keys())
print(duration_df)
```

```python
import matplotlib.pyplot as plt
# plotting
duration_df.plot(kind='line',marker=0)
plt.xlabel('Years')
plt.ylabel('duration')
plt.title('Netflix Movie Durations 2011-2020')
```

```python
path = r'C:\Users\user\Desktop\new individual pject\netflix_data.csv'
netflix_df = pd.read_csv(path)
print(netflix_df.head(5))
```

```python
netflix_df_movies_only = netflix_df.loc[netflix_df['type'].isin(['Movie'])]


netflix_movie_col_subset = netflix_df_movies_only[['title','country','genre','release_year','duration']]

print(netflix_movie_col_subset.head(5))
```

```python
import matplotlib.pyplot as plt
netflix_movie_col_subset.plot.scatter(x='release_year', y='duration')
plt.xlabel('release_year')
plt.ylabel('duration')
plt.title('Movie Duration By Year of Release')
plt.show()
```
